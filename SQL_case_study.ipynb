```python
#Q1
#/* Q1: Some of the facilities charge a fee to members, but some do not.
#Write a SQL query to produce a list of the names of the facilities that do. */
pd.read_sql_query('SELECT name FROM `Facilities` WHERE membercost > 0', engine)
```

```python
#/* Q2: How many facilities do not charge a fee to members? */
pd.read_sql_query('SELECT COUNT(*) FROM `Facilities` WHERE membercost = 0', engine)
```

```python
"""
/* Q3: Write an SQL query to show a list of facilities that charge a fee to members,
where the fee is less than 20% of the facility's monthly maintenance cost.
Return the facid, facility name, member cost, and monthly maintenance of the
facilities in question. */
"""
pd.read_sql_query("""SELECT facid, name, membercost, monthlymaintenance
FROM Facilities
WHERE membercost < (monthlymaintenance * 0.2)
AND membercost > 0
""", engine)
```

```python
"""/* Q4: Write an SQL query to retrieve the details of facilities with ID 1 and 5.
Try writing the query without using the OR operator. */
"""

pd.read_sql_query('SELECT * FROM Facilities WHERE facid IN (1,5)', engine)
```

```python
"""/* Q5: Produce a list of facilities, with each labelled as
'cheap' or 'expensive', depending on if their monthly maintenance cost is
more than $100. Return the name and monthly maintenance of the facilities
in question. */
"""

pd.read_sql_query("""SELECT name, monthlymaintenance, 
	CASE WHEN monthlymaintenance > 100 THEN 'expensive'
	ELSE 'cheap' END AS cost_label
FROM Facilities
""", engine)
```

```python
"""/* Q6: You'd like to get the first and last name of the last member(s)
who signed up. Try not to use the LIMIT clause for your solution. */
"""

pd.read_sql_query("""SELECT firstname, surname, joindate
FROM Members
ORDER BY joindate DESC
""", engine)
```

```python
"""/* Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name. */
"""

pd.read_sql_query("""SELECT f.name AS facility_name, 
	firstname||' '||surname AS member_name
FROM Bookings AS b
INNER JOIN (SELECT facid, name FROM Facilities WHERE name LIKE 'Tennis%') as f
ON b.facid = f.facid
INNER JOIN Members AS m
ON b.memid = m.memid
GROUP BY facility_name, member_name
""", engine)
```

```python
"""
/* Q8: Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries. */
"""

pd.read_sql_query("""SELECT f.name as facility_name, 
	m.firstname||' '||m.surname AS member_name,
	CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END AS cost
FROM Bookings AS b
LEFT JOIN Facilities as f
ON b.facid = f.facid
LEFT JOIN Members as m
ON m.memid = b.memid
WHERE b.starttime LIKE '2012-09-14%'
AND (CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) > 30
ORDER BY cost DESC
""",engine)
```

```python
"""/* Q9: This time, produce the same result as in Q8, but using a subquery. */
"""

pd.read_sql_query("""SELECT facility_name, member_name, cost
FROM (
	SELECT b.memid, f.name AS facility_name, 
		m.firstname||' '||m.surname AS member_name, b.starttime,
		CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END AS cost
	FROM Bookings AS b
	LEFT JOIN Facilities AS f
	USING (facid)
	LEFT JOIN Members AS m
	USING (memid)) AS subquery
WHERE starttime LIKE '2012-09-14%'
AND cost > 30
ORDER BY cost DESC
""", engine)
```

# Now for Part 2 (Questions 10-13)

```python
"""/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */
"""
# revenue = sum of price_per_booking for all bookings for that facility
pd.read_sql_query("""SELECT facility_name, total_revenue
    FROM (SELECT f.name AS facility_name, 
        SUM(CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) AS total_revenue
        FROM Bookings AS b
        INNER JOIN Facilities as f
        USING(facid)
        GROUP BY facility_name)
    WHERE total_revenue < 1000
    ORDER BY total_revenue    
    """, engine)
```

```python
"""
/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */
"""

pd.read_sql_query("""SELECT m1.surname||', '||m1.firstname AS member_name,
     m2.surname||', '||m2.firstname AS recommended_by
     FROM Members AS m1
     LEFT JOIN Members AS m2
     ON m2.memid = m1.recommendedby
     WHERE member_name NOT LIKE '%GUEST%'
     ORDER BY member_name
""", engine)
```

```python
# The phrasing of Q12 is unclear... What does usage mean? "by member" suggests individual members should be displayed, 
#but the next question (Q13) makes it sound like it should be total member uses by all members. 
"""/* Q12: Find the facilities with their usage by member, but not guests */
"""

pd.read_sql_query("""SELECT f.name AS facility,
    m.surname||', '||m.firstname AS used_by_member,
    COUNT(*) AS times_used_by_member
     FROM Facilities AS f
     LEFT JOIN Bookings AS b
     USING(facid)
     LEFT JOIN Members AS m
     USING(memid)
     WHERE used_by_member NOT LIKE '%GUEST%'
     GROUP BY facility, used_by_member
     ORDER BY facility, used_by_member
     
""", engine)
```

```python
"""
/* Q13: Find the facilities usage by month, but not guests */
"""

pd.read_sql_query("""SELECT f.name AS facility,
    STRFTIME('%m', b.starttime) AS month,
    COUNT(*) AS total_bookings_by_members_each_month
     FROM Facilities AS f
     LEFT JOIN Bookings AS b
     USING(facid)
     LEFT JOIN Members AS m
     USING(memid)
     WHERE m.firstname NOT LIKE '%GUEST%'
     GROUP BY facility, month
     ORDER BY facility, month
""", engine)
```
